```python
import pandas as pd

def formatar(valor):
    return "R${:,.2f}".format(valor) # formatado

tabela = pd.read_excel("Vendas.xlsx")
# tabela["Vendas"] = tabela["Vendas"].apply(formatar)
display(tabela)
```

```python
imposto = 0.1
tabela["Imposto"] = tabela["Vendas"] * imposto
tabela["Vendas"] = tabela["Vendas"].apply(formatar)
display(tabela)
```
