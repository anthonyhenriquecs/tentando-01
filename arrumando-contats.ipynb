```python
import pandas as pd

# Carrega a tabela
tabela = pd.read_csv("contacts.csv")

# Função para corrigir os números de telefone
def corrigir_numero(numero):
    if pd.notna(numero):
        numero_str = str(numero)
        if not numero_str.startswith("+5561"):
            numero = "+5561" + numero_str
    return numero

# Aplica a função para corrigir os números de telefone na coluna desejada
tabela['Celular2'] = tabela['Celular2'].apply(corrigir_numero)

# Salva a tabela atualizada
tabela.to_csv("contacts.csv", index=False)
display(tabela)

```
